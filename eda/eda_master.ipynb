```python
# =====================================================
# ðŸ““ Notebook: eda_master.ipynb
# ðŸŽ¯ Exploratory Data Analysis (EDA) Maestro - ComparaciÃ³n entre datasets:
#     - No Social (sin variables sociales)
#     - Imputed Full (con imputaciÃ³n y variables sociales)
# =====================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ===========================
# 0. ConfiguraciÃ³n general
# ===========================
sns.set(style="whitegrid", palette="muted", font_scale=1.2)
plt.rcParams["figure.figsize"] = (8, 5)

# ===========================
# 1. Cargar datasets
# ===========================
df_no_social = pd.read_csv("../data/processed/places_no_social_clean.csv")
df_imputed = pd.read_csv("../data/processed/places_imputed_full_clean.csv")

print("ðŸ“˜ No Social:", df_no_social.shape)
print("ðŸ“— Imputed Full:", df_imputed.shape)

# ===========================
# 2. EstadÃ­sticas descriptivas
# ===========================
with pd.option_context("display.float_format", "{:.2f}".format):
    print("\n--- Descriptivas (No Social) ---")
    display(df_no_social.describe())
    print("\n--- Descriptivas (Imputed Full) ---")
    display(df_imputed.describe())

# ===========================
# 3. DistribuciÃ³n de targets
# ===========================
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(df_no_social["depression_crudeprev"], bins=30, kde=True, ax=axes[0], color="steelblue")
axes[0].set_title("DistribuciÃ³n de DepresiÃ³n (No Social)")
axes[0].set_xlabel("Depression (Crude Prevalence)")

sns.histplot(df_imputed["depression_crudeprev"], bins=30, kde=True, ax=axes[1], color="seagreen")
axes[1].set_title("DistribuciÃ³n de DepresiÃ³n (Imputed Full)")
axes[1].set_xlabel("Depression (Crude Prevalence)")

plt.tight_layout()
plt.show()

# Frequent Mental Distress (MHLTH)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(df_no_social["mhlth_crudeprev"], bins=30, kde=True, ax=axes[0], color="steelblue")
axes[0].set_title("Frequent Mental Distress (No Social)")
axes[0].set_xlabel("MHLTH Crude Prevalence")

sns.histplot(df_imputed["mhlth_crudeprev"], bins=30, kde=True, ax=axes[1], color="seagreen")
axes[1].set_title("Frequent Mental Distress (Imputed Full)")
axes[1].set_xlabel("MHLTH Crude Prevalence")

plt.tight_layout()
plt.show()

# ===========================
# 4. Outliers principales
# ===========================
features_outliers = [
    "depression_crudeprev",
    "obesity_crudeprev",
    "lpa_crudeprev",          # low physical activity
    "mhlth_crudeprev"
]

for feat in features_outliers:
    if feat in df_no_social.columns:
        fig, axes = plt.subplots(1, 2, figsize=(14, 5))
        sns.boxplot(y=df_no_social[feat], ax=axes[0], color="steelblue")
        axes[0].set_title(f"{feat} (No Social)")

        sns.boxplot(y=df_imputed[feat], ax=axes[1], color="seagreen")
        axes[1].set_title(f"{feat} (Imputed Full)")

        plt.tight_layout()
        plt.show()

# ===========================
# 5. Correlaciones con depresiÃ³n
# ===========================
num_cols_no_social = df_no_social.select_dtypes(include="number")
num_cols_imputed = df_imputed.select_dtypes(include="number")

corr_no_social = num_cols_no_social.corr()["depression_crudeprev"].sort_values(ascending=False)
corr_imputed = num_cols_imputed.corr()["depression_crudeprev"].sort_values(ascending=False)

print("\nðŸ”¹ Top correlaciones (No Social):")
display(corr_no_social.head(10))

print("\nðŸ”¹ Top correlaciones (Imputed Full):")
display(corr_imputed.head(10))

# Guardar resumen comparativo
summary_corr = pd.DataFrame({
    "no_social": corr_no_social,
    "imputed_full": corr_imputed
}).fillna(0)

summary_corr.to_csv("../data/interim/corr_comparison_master.csv")
print("\nðŸ“Š Guardado resumen comparativo en ../data/interim/corr_comparison_master.csv")

# ===========================
# 6. Heatmaps comparativos
# ===========================
plt.figure(figsize=(12, 8))
sns.heatmap(num_cols_no_social.corr().iloc[:20, :20], cmap="coolwarm", center=0)
plt.title("Heatmap correlaciones (No Social)")
plt.show()

plt.figure(figsize=(12, 8))
sns.heatmap(num_cols_imputed.corr().iloc[:20, :20], cmap="coolwarm", center=0)
plt.title("Heatmap correlaciones (Imputed Full)")
plt.show()

# ===========================
# 7. Scatterplots clave
# ===========================
pairs = [
    ("obesity_crudeprev", "depression_crudeprev"),
    ("lpa_crudeprev", "depression_crudeprev"),
    ("mhlth_crudeprev", "depression_crudeprev"),
    ("ghlth_crudeprev", "depression_crudeprev")  # general health
]

for x, y in pairs:
    if x in df_no_social.columns and y in df_no_social.columns:
        fig, axes = plt.subplots(1, 2, figsize=(14, 5))
        sns.scatterplot(x=df_no_social[x], y=df_no_social[y], ax=axes[0], color="steelblue", alpha=0.6)
        axes[0].set_title(f"{y} vs {x} (No Social)")

        sns.scatterplot(x=df_imputed[x], y=df_imputed[y], ax=axes[1], color="seagreen", alpha=0.6)
        axes[1].set_title(f"{y} vs {x} (Imputed Full)")

        plt.tight_layout()
        plt.show()

# ===========================
# 8. Conclusiones generales
# ===========================
print("""
ðŸ“Œ Conclusiones del anÃ¡lisis maestro:
- 'mhlth_crudeprev' (Frequent Mental Distress) muestra la correlaciÃ³n mÃ¡s alta con 'depression_crudeprev'.
- 'obesity_crudeprev' y 'lpa_crudeprev' (actividad fÃ­sica baja) tambiÃ©n correlacionan positivamente.
- 'ghlth_crudeprev' (salud general) presenta correlaciÃ³n negativa: peor salud â†’ mÃ¡s depresiÃ³n.
- Las variables sociales del dataset imputado podrÃ­an explicar variabilidad residual.
- Ambos datasets mantienen una estructura coherente â†’ excelente base para modelado.
""")
```
